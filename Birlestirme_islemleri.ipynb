## Birleştirme(Join) İşlemleri

```python
import numpy as np
import pandas as pd
m=np.random.randint(1,30,size=(5,3))
df=pd.DataFrame(m,columns=["var1","var2","var3"])
df
```

```python
df2=df+10
```

```python
df2
```

```python
pd.concat([df,df2]) #indexlerde bir anormallik var bunu nasıl düzeltebilirim?
```

```python
pd.concat([df,df2],ignore_index=True) #çözümü ignore_index ile yaptık 
```

```python
df.columns
```

```python
df2.columns=["var1","var2","değişken1"]
```

```python
df2
```

burada bi eksiklik var birleştirme yaptığım zaman NaN değeri dönüyor çünkü df de değişken1 adlı columns yok peki biz bunu nasıl düzelteceğiz

```python
pd.concat([df,df2]) 
```

```python
pd.concat([df,df2],join="inner") #ortak kesişimleri olanları birleştirtik

#farklı değişkenlerden oluşan veri setlerini birleştirmek mantıksızdır zaten.
```

## İleri Birleştirme İşlemleri

```python
import pandas as pd
```

```python
#bire bir birleştirme(tüm elemanların 2 veri setinde de yer alması durumu)
```

```python
df1=pd.DataFrame({
    "çalışanlar": ["ali","veli","ayşe","fatma"],
    "grup": ["muhasebe","mühendislik","mühendislik","ik"]
})
df1
```

```python
df2=pd.DataFrame({
    "çalışanlar":["ayşe","ali","veli","fatma"],
    "ilk_giriş":[2010,2009,2014,2019]
})
df2
```

```python
pd.merge(df1,df2)
```

```python
pd.merge(df1,df2, on="çalışanlar")
```

```python
#çoktan teke
```

```python
df3=pd.merge(df1,df2)
```

```python
df3
```

```python
df4=pd.DataFrame(
    {
         "grup": ["muhasebe","mühendislik","ik"],
        "müdür":["caner","mustafa","ebru"]
        
    }
)
df4
```

```python
pd.merge(df3,df4)
```

```python
#çoktan çoka
```

```python
df5=pd.DataFrame({
     "grup": ["muhasebe","muhasebe","mühendislik","mühendislik","ik","ik"],
    "yetenekler":["matematik","excel","kodlama","linux","excel","yönetim"]
})
df5
```

```python
pd.merge(df1,df5)
```

```python
##gruplama işlemleri
```

```python
df=pd.DataFrame({
    "gruplar":["a","b","c","a","b","c"],
    "veri":[10,11,52,23,43,55]},columns=["gruplar","veri"]
)
df
```

```python
df.groupby("gruplar")
```

```python
df.groupby("gruplar").mean()
```

```python
df.groupby("gruplar").sum()
```
